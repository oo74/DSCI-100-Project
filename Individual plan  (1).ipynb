### Played Hours Distribution

```r
ggplot(players, aes(x = played_hours)) +
  geom_histogram(binwidth = 5, fill = "blue", color = "black", alpha = 0.7) +
  labs(title = "Distribution of Played Hours", 
       x = "Played Hours", 
       y = "Count") +
theme(text = element_text(size = 16))
```

Most players have a short played hours, a few players have a long played hours.

### Age distribution

```r
ggplot(players, aes(x = Age)) +
  geom_histogram(binwidth = 2, fill = "green", color = "black", alpha = 0.7) +
  labs(title = "Distribution of Player Age", 
       x = "Age", 
       y = "Count") +
theme(text = element_text(size = 16))
```

The majority of players are young adults and teenagers, most of them are between 15 and 30 years old.

### The relationship between subscription rate and experience level

```r
ggplot(players, aes(x = experience, fill = subscribe)) +
  geom_bar(stat = "count") +
  labs(title = "Subscription Rate by Experience Level", 
       xlab = "Experience Level", 
       ylab = "Count") +
  scale_fill_manual(values = c("blue", "green"), name = "Subscribed") +
theme(text = element_text(size = 16))
```

More experienced players tend to subscribe newsletters of the game. (Pro, Veteran and Regular)

### The relationship between subscription rate and gender

```r
ggplot(players, aes(x = gender, fill = subscribe)) +
  geom_bar(stat = "count") +
  labs(title = "Subscription Rate by Gender", 
       xlab = "Gender", 
       ylab = "Count") +
  scale_fill_manual(values = c("orange", "purple"), name = "Subscribed") +
theme(text = element_text(size = 16))
```

Male is more likely to subscribe the newsletters.

### The relationship between age and played hours

```r
ggplot(players, aes(x = Age, y = played_hours)) +
geom_point(aes(color = subscribe)) +
labs(title = "The relationship between age and played hours",
     xlab = "Age",
     ylab = "played hours") +
theme(text = element_text(size = 16))
```

The people who are younger(between 10 and 30 years old) and have more played hours are more likely to subscribe newsletters of the game.

## Methods and plan
To predict whether the player will subscribe the newsletter of the game, I will choose the K-Nearest Neighbors (KNN) classification method.

KNN is a supervised learning algorithm that is particularly useful for classification problems where patterns and similiaries in the data is important.

#### Reasons for choosing KNN as methods: 
KNN does not assume any underlying distribution of the data, making it useful when relationships between variables are unknown or complex. KNN can capture non-linear patterns in the data. KNN classifies new data points by looking at the majority class among its nearest neighbors, making the decision process transparent and easy to understand.

#### Potentical limitions: 
KNN may be sensitive to the irrelevant data, which can affect the calculations of distance and lead to a poor performance.
